```python
import pandas as pd
df = pd.read_csv('/kaggle/input/all-the-unicorns-in-the-world/List of Unicorns in the World.csv')

# Step 1 remove israel

country_to_del = "Israel"
df_clean = df[df["Country"] != country_to_del].copy()

# Step 2 count startups
startup_counts = df_clean["Country"].value_counts()

# Step 3 calc quantiles
quantiles = startup_counts.quantile([0.99, 0.95, 0.90]) 

df_clean["Count"] = df_clean["Country"].map(startup_counts)
df_clean["Quantiles"] = df_clean["Count"].map(quantiles)
startup_counts_sorted = startup_counts.sort_values(ascending=False)

print(df_clean.head()) 
```

```python
# here I will make data look fancy

# Step 0 get matplot and numpy to create graphs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data into a DataFrame
df = pd.read_csv('/kaggle/input/all-the-unicorns-in-the-world/List of Unicorns in the World.csv')

# Step 1: Clean the data
country_to_del = "Israel"
df_clean = df[df["Country"] != country_to_del].copy()

# Step 2: Link the data
unique_industry = df_clean["Industry"].unique()
count_startups = df_clean["Industry"].value_counts()

# Step 3: Set plot variables
x = unique_industry
y = count_startups

# Function to add value labels
def addlabels(x, y):
    for i in range(len(x)):
        plt.text(y.iloc[i], i, y.iloc[i],ha = 'left')

# Step 4: Display fully linked data in graph

plt.figure(figsize = (10,5))
plt.barh(x, y, color='lightblue', linewidth=2, edgecolor="red")
plt.title('Total startups per industry')
plt.xlabel('Number of startups')    
plt.ylabel('Industries')
addlabels(x, y)
plt.show()







```

```python
# here I want to check in what year startups were founded, also I want to check data that there is no years like9999

# Step 0 get libs and data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('/kaggle/input/all-the-unicorns-in-the-world/List of Unicorns in the World.csv')

# Step 1: Clean the data

country_to_del = "Israel"
df_clean = df[df["Country"] != country_to_del].copy()

# Step 2 new column with years def function which returns year out of the date 


df_clean["Year"] = pd.DatetimeIndex (df_clean["Date Joined"]).year
df_clean["Month"] = pd.DatetimeIndex (df_clean["Date Joined"]).month


unique_year = df_clean["Year"].astype(int).unique()
count_years = df_clean["Year"].value_counts().sort_index()

# Step 3: Set plot variables

x = unique_year
y = count_years

plt.barh(x,y)
plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.show()






```
